```python
# ====================
# === INISIALISASI ===
# ====================
produk_roti = []
riwayat_pembelian = []
total_pemasukan = 0

# ====================
# === UTILITAS INPUT ==
# ====================
def input_angka(pesan):
    while True:
        try:
            return int(input(pesan))
        except ValueError:
            print("‚ùå Masukkan angka yang valid.")

def tampilkan_stok():
    if not produk_roti:
        print("‚ùå Stok masih kosong.")
        return False
    df = pd.DataFrame(produk_roti)
    try:
        display(df)
    except:
        print("\nüìã Daftar Stok Roti")
        print(df.to_string(index=False))
    return True

# =====================
# === FITUR PENJUAL ===
# =====================
def tambah_produk():
    while True:
        print("\nüì¶ Tambah Produk Roti")
        nama = input("Nama Roti: ")
        harga = input_angka("Harga Roti (Rp): ")
        stok = input_angka("Jumlah Stok: ")

        produk_roti.append({"Nama": nama, "Harga": harga, "Stok": stok})
        lagi = input("Tambah produk lagi? (y/n): ").lower()
        if lagi != 'y':
            break

def lihat_pemasukan(pemasukan):
    print(f"\nüí∞ Total Pemasukan Saat Ini: Rp{pemasukan}")

def lihat_riwayat():
    if not riwayat_pembelian:
        print("\nüïí Belum ada transaksi.")
        return
    print("\nüßæ Riwayat Pembelian:")
    for i, data in enumerate(riwayat_pembelian, 1):
        print(f"{i}. {data['Nama Pembeli']} - Total: Rp{data['Total Belanja']}")
        for item in data["Detail"]:
            print(f"   - {item['Jumlah']} x {item['Nama']} = Rp{item['Subtotal']}")

def menu_penjual():
    global total_pemasukan
    while True:
        print("""
üì¶=== Menu Penjual ===
1. Tambah Roti ke Stok
2. Lihat Stok Roti
3. Lihat Total Pemasukan
4. Lihat Riwayat Pembelian
5. Kembali
""")
        pilihan = input("Pilih menu (1-5): ")

        if pilihan == '1':
            tambah_produk()
        elif pilihan == '2':
            while True:
                ada = tampilkan_stok()
                if not ada:
                    break
                print("""
üìã Apa selanjutnya?
1. Tambah Produk Baru
2. Lihat Stok Lagi
3. Kembali ke Menu Penjual
""")
                next_action = input("Pilih menu (1-3): ")
                if next_action == '1':
                    tambah_produk()
                elif next_action == '2':
                    continue
                elif next_action == '3':
                    break
                else:
                    print("‚ùå Pilihan tidak valid.")
        elif pilihan == '3':
            lihat_pemasukan(total_pemasukan)
        elif pilihan == '4':
            lihat_riwayat()
        elif pilihan == '5':
            break
        else:
            print("‚ùå Pilihan tidak valid.")

# =====================
# === FITUR PEMBELI ===
# =====================
def proses_pembelian():
    global total_pemasukan
    if not produk_roti:
        print("‚ùå Belum ada roti tersedia.")
        return

    nama = input("\nüßç Nama Pembeli: ")
    keranjang = []
    total = 0

    while True:
        print("""
üõí=== Menu Pembeli ===
1. Lihat Stok Roti
2. Beli Roti
3. Selesai dan Bayar
""")
        pilihan = input("Pilih menu (1-3): ")

        if pilihan == '1':
            tampilkan_stok()

        elif pilihan == '2':
            tampilkan_stok()
            nama_roti = input("Nama roti yang ingin dibeli: ")
            jumlah = input_angka("Jumlah yang ingin dibeli: ")

            for roti in produk_roti:
                if roti["Nama"].lower() == nama_roti.lower():
                    if roti["Stok"] >= jumlah:
                        subtotal = roti["Harga"] * jumlah
                        keranjang.append({
                            "Nama": roti["Nama"],
                            "Harga": roti["Harga"],
                            "Jumlah": jumlah,
                            "Subtotal": subtotal
                        })
                        total += subtotal
                        roti["Stok"] -= jumlah
                        print(f"‚úÖ Ditambahkan: {jumlah} x {roti['Nama']} = Rp{subtotal}")
                    else:
                        print("‚ùå Stok tidak cukup.")
                    break
            else:
                print("‚ùå Roti tidak ditemukan.")

        elif pilihan == '3':
            break
        else:
            print("‚ùå Pilihan tidak valid.")

    if keranjang:
        print("\nüßæ Ringkasan Belanja:")
        for item in keranjang:
            print(f"- {item['Jumlah']} x {item['Nama']} = Rp{item['Subtotal']}")
        print(f"Total Bayar: Rp{total}")
        total_pemasukan += total
        riwayat_pembelian.append({
            "Nama Pembeli": nama,
            "Total Belanja": total,
            "Detail": keranjang
        })
        print("‚úÖ Terima kasih atas pembelian Anda!")
    else:
        print("‚ùó Tidak ada item yang dibeli.")

# ====================
# === MENU UTAMA =====
# ====================
def menu_utama():
    while True:
        print("""
====== TOKO ROTI ======
1. Login sebagai Pembeli
2. Login sebagai Penjual
3. Keluar
""")
        pilihan = input("Pilih menu (1-3): ")

        if pilihan == '1':
            proses_pembelian()
        elif pilihan == '2':
            menu_penjual()
        elif pilihan == '3':
            print("üëã ARIGATOU GOZAIMASU!")
            break
        else:
            print("‚ùå Pilihan tidak valid.")

# Mulai Program
menu_utama()
```
