```python
import pandas as pd

# ====================
# === INISIALISASI ===
# ====================
produk_roti = []
riwayat_pembelian = []
total_pemasukan = 0

# ====================
# === FUNGSI PENJUAL ==
# ====================
def tambah_produk(): #untuk tambah produk
    while True:
        print("\nüì¶ Tambah Produk Baru") #data yang harus diinput untuk stok
        nama = input("Nama Roti: ")
        harga = int(input("Harga (Rp): "))
        stok = int(input("Jumlah Stok: "))

        produk_roti.append({
            "Nama": nama,
            "Harga": harga,
            "Stok": stok
        })

        lagi = input("Tambah produk lagi? (y/n): ").lower() #opsi jika ingin/tidak ingin menambahkan produk
        if lagi != 'y':
            break

def lihat_stok(): #untuk melihat stok yang ada
    if produk_roti:
        df = pd.DataFrame(produk_roti)
        print("\nüìã Daftar Stok Roti")
        try:
            display(df)  # Cocok untuk Google Colab / Jupyter
        except:
            print(df.to_string(index=False))  # Terminal / VSCode
    else:
        print("‚ùå Stok kosong. Tambah dulu produknya.")

def lihat_pemasukan(): #untuk melihat pemasukan
    print(f"\nüí∞ Total Pemasukan: Rp{total_pemasukan}")

def lihat_riwayat(): #untuk melihat riwayat pembelian
    if not riwayat_pembelian:
        print("\nüïí Belum ada transaksi.")
        return

    print("\nüßæ Riwayat Pembelian:")
    for i, r in enumerate(riwayat_pembelian, 1):
        print(f"{i}. Pembeli: {r['Nama Pembeli']} - Total: Rp{r['Total Belanja']}")
        for item in r["Detail"]:
            print(f"   - {item['Jumlah']} x {item['Nama']} = Rp{item['Subtotal']}")

def menu_penjual(): #untuk menampilkan menu penjual
    while True:
        print("""
üì¶=== Menu Penjual ===
1. Tambah Roti ke Stok
2. Lihat Stok Roti
3. Lihat Total Pemasukan
4. Lihat Riwayat Pembelian
5. Kembali ke Menu Utama
""")
        pilihan = input("Pilih menu (1-5): ")

        if pilihan == '1':
            tambah_produk()

        elif pilihan == '2':
            while True:
                lihat_stok()
                print("""
üìã Apa yang ingin Anda lakukan selanjutnya?
1. Lihat Stok Lagi
2. Tambah Produk Roti
3. Kembali ke Menu Penjual
""")
                lanjut = input("Pilih menu (1-3): ")

                if lanjut == '1':
                    continue
                elif lanjut == '2':
                    tambah_produk()
                elif lanjut == '3':
                    break
                else:
                    print("Pilihan tidak valid.")

        elif pilihan == '3':
            lihat_pemasukan()

        elif pilihan == '4':
            lihat_riwayat()

        elif pilihan == '5':
            break

        else:
            print("Pilihan tidak valid.")

# ====================
# === FUNGSI PEMBELI =
# ====================
def pembeli():
    global total_pemasukan
    if not produk_roti:
        print("\n‚ùå Belum ada roti tersedia.")
        return

    nama_pembeli = input("\nMasukkan nama Anda: ")
    keranjang = []
    total = 0

    while True:
        print("""
üõí=== Menu Pembeli ===
1. Lihat Stok Roti
2. Beli Roti
3. Selesai dan Bayar
""")
        pilihan = input("Pilih menu (1-3): ")

        if pilihan == '1':
            lihat_stok()

        elif pilihan == '2':
            lihat_stok()
            nama_roti = input("Masukkan nama roti yang ingin dibeli: ")
            jumlah = int(input("Jumlah yang ingin dibeli: "))

            for roti in produk_roti:
                if roti["Nama"].lower() == nama_roti.lower():
                    if roti["Stok"] >= jumlah:
                        subtotal = roti["Harga"] * jumlah
                        keranjang.append({
                            "Nama": roti["Nama"],
                            "Harga": roti["Harga"],
                            "Jumlah": jumlah,
                            "Subtotal": subtotal
                        })
                        total += subtotal
                        roti["Stok"] -= jumlah
                        print(f"‚úÖ Ditambahkan: {jumlah} x {roti['Nama']} = Rp{subtotal}")
                    else:
                        print("‚ùå Stok tidak cukup.")
                    break
            else:
                print("‚ùå Roti tidak ditemukan.")

        elif pilihan == '3':
            break
        else:
            print("Pilihan tidak valid.")

    if keranjang:
        print("\nüßæ Ringkasan Belanja:")
        for item in keranjang:
            print(f"- {item['Jumlah']} x {item['Nama']} = Rp{item['Subtotal']}")
        print(f"Total yang harus dibayar: Rp{total}")
        total_pemasukan += total
        riwayat_pembelian.append({
            "Nama Pembeli": nama_pembeli,
            "Total Belanja": total,
            "Detail": keranjang
        })
        print("‚úÖ Terima kasih atas pembelian Anda!")
    else:
        print("Belum ada pembelian yang dilakukan.")

# ====================
# === PROGRAM UTAMA ==
# ====================
while True:
    print("""
====== TOKO ROTI ======
1. Login sebagai Pembeli
2. Login sebagai Penjual
3. Keluar
""")
    menu = input("Pilih menu (1-3): ")

    if menu == '1':
        pembeli()
    elif menu == '2':
        menu_penjual()
    elif menu == '3':
        print("üëã Terima kasih telah menggunakan aplikasi Toko Roti!")
        break
    else:
        print("Pilihan tidak valid.")
```
